## function to convert the discounted_price column

```python
def extract_integer(value):
    if pd.isna(value):
        return value
    elif isinstance(value, str):
        integer_part = value.split('.')[0]
        return int(integer_part)
    elif isinstance(value, float):
        return int(value)
    else:
        return None

# Use .loc[] to modify the original DataFrame
df1.loc[:, 'discounted_price'] = df1.loc[:,'discounted_price'].apply(extract_integer)
print(df1)
```

```python
scaler = StandardScaler()
scaler.fit(df5.drop('rating',axis=1))
scaled = scaler.transform(df5.drop('rating',axis=1))
scaleddf = pd.DataFrame(scaled,columns=df5.columns[:-1])
scaleddf
```

## model building

```python
X = scaleddf
y = df5['rating']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
err_rate = []

for i in range(1,40,2):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    err_rate.append(np.mean(pred_i != y_test))
```

## plot graph

```python
plt.figure(figsize=(20,10))
plt.plot(range(1,40,2),err_rate,color='blue',linestyle='dotted',marker='o',markerfacecolor='red',markersize=8)
plt.title = 'K Values VS Error Rates'
plt.xlabel = 'K Value'
plt.ylabel= 'Error Rate'
plt.show()
```

## run model using best k value

```python
knn = KNeighborsClassifier(n_neighbors=31)
knn.fit(X,y)
```

```python
pred = knn.predict(X_test)
pred
```
